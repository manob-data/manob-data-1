```python
import pandas as pd
df_sales= pd.read_excel("linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=[12,4])
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="red")
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=[12,4])
plt.plot(df_sales["Dishwasher"], df_sales["Washing Machine"])
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=[12,4])
plt.plot(df_sales["Quarter"], df_sales["Fridge"])
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="blue", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="green", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="red", label="Washing Machine")

plt.title("Product Sales")
plt.ylabel("Revenue(mln $)")
plt.xlabel("Financial Quarter")    


plt.legend()

plt.show()
```
